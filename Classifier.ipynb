```python
xtrain=train.iloc[:,:-1]
ytrain=train.iloc[:,[11]]
xtest=test.iloc[:,:-1]
ytest=test.iloc[:,[11]]
```

```python
train_stats = train.describe()
#sns.pairplot(train_stats[train_stats.columns], diag_kind = "kde")
```

```python
scaler=preprocessing.StandardScaler()
scaler.fit(xtrain)
scaler.fit(xtest)
```

```python
xtrain=scaler.transform(xtrain)
xtrain
```

```python
xtest=scaler.transform(xtest)
xtest
```

```python
clf=svm.SVC()
clf.fit(xtrain,ytrain)
acc=clf.score(xtest,ytest)
print(acc)
train_preds=clf.predict(xtrain)
test_preds=clf.predict(xtest)
print("cost of training model\n",metrics.confusion_matrix(ytrain,train_preds))
print("cost of training model\n",metrics.classification_report(ytrain,train_preds))
print("cost of testing model\n",metrics.confusion_matrix(ytest,test_preds))
print("cost of training model\n",metrics.classification_report(ytrain,train_preds))
```

```python
clf=RandomForestClassifier()
clf.fit(xtrain,ytrain)
acc=clf.score(xtest,ytest)
print(acc)
train_preds=clf.predict(xtrain)
test_preds=clf.predict(xtest)
print("cost of training model\n",metrics.confusion_matrix(ytrain,train_preds))
print("cost of training model\n",metrics.classification_report(ytrain,train_preds))
print("cost of testing model\n",metrics.confusion_matrix(ytest,test_preds))
print("cost of training model\n",metrics.classification_report(ytrain,train_preds))
```
